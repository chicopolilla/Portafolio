```python
numerics = breast_cancer._get_numeric_data()

numerics.head(5)
```

```python
numerics = numerics.drop(['id','time'], axis=1)
```

```python
# standarize de data 

from sklearn.preprocessing import StandardScaler
transformer = StandardScaler().fit(numerics)
scaled = transformer.transform(numerics)
scaled = pd.DataFrame(scaled)
scaled.columns = numerics.columns 
scaled.head()
```

```python
from sklearn.decomposition import PCA
from sklearn import preprocessing
pca = PCA()
pca.fit(scaled)
```

```python
pca.components_
```

```python
variations = pca.explained_variance_
variations = pca.explained_variance_ratio_*100
variations = np.round(variations, decimals=1)
variations
```

```python
plot_labels = ['PC'+str(i) for i in range(1,len(variations)+1)]
?plot_labels
```

```python
from matplotlib.pyplot import figure
figure(figsize=(20,5))
plt.bar(x = plot_labels, height=variations)
plt.xlabel('Principal Components')
plt.ylabel('Variance Explained')
plt.title('PCA Scree Plot')
```

```python
data = pd.DataFrame(pca.components_)
data = data[[0,1,2]]
data.columns = ['PC1', 'PC2', 'PC3']
data.head()
```

One common approach to determining the number of PCs to retain is to use the scree plot. This is a plot of the eigenvalues associated with each PC, sorted from largest to smallest. The point where the eigenvalues "level off" or start to become relatively small can be used as a cutoff point for retaining PCs. This point can be somewhat subjective and may depend on the specific context and goals of the analysis.

Another approach is to use a rule of thumb such as retaining enough PCs to explain a certain percentage of the total variability in the data, such as 80% or 90%.
