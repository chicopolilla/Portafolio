```python
url = 'https://www.billboard.com/charts/hot-100/'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Find all li elements with class "o-chart-results-list__item"
list_items = soup.find_all('li', class_='o-chart-results-list__item')

# Create an empty list to store the rank numbers
rank_numbers = []

# Iterate over each li element and find the span element with class "c-label"
# which contains the rank number
for item in list_items:
    span = item.find('span', class_='c-label')
    if span is not None:
        text = span.get_text().strip()
        rank_numbers.append(text)

# Create a DataFrame with the rank numbers and print it
numbers = pd.DataFrame(rank_numbers, columns=['Rank'])
print(numbers)
```

```python
numbers = numbers.drop([0])
print(numbers)
```

```python
numbers = numbers[numbers['Rank'] != 'NEW']
print(numbers)
```

```python
start = 0
end = 4
total_rows = len(numbers)

# crea un dataframe vacío para almacenar las filas conservadas
numbers_filtered = pd.DataFrame(columns=numbers.columns)

# mientras queden filas por procesar
while end <= total_rows:
    # agrega las filas a numbers_filtered
    numbers_filtered = pd.concat([numbers_filtered, numbers[start:end]], ignore_index=True)
    # actualiza los indices de inicio y fin
    start += 8
    end += 8

# agrega las últimas filas si no fueron suficientes para completar una serie de 8 filas
if start < total_rows:
    numbers_filtered = pd.concat([numbers_filtered, numbers[start:]], ignore_index=True)

# imprime el resultado
print(numbers_filtered)
```
