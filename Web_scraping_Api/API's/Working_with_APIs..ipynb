Following the class example, create a function that returns the price, names of origin and arrival airports and the name of the company. Do it for all the flights between two dates that cost the same.

```python
def get_flight_prices(start_date, end_date, origin_airport, destination_airport, api_key):
    url = 'https://partners.api.skyscanner.net/apiservices/v3/browsequotes'
    headers = {'prtl6749387986743898559646983194': api_key}
    params = {
        'country': 'US',
        'currency': 'USD',
        'locale': 'en-US',
        'originplace': origin_airport,
        'destinationplace': destination_airport,
        'outboundpartialdate': start_date,
        'inboundpartialdate': end_date,
    }
    response = requests.get(url, headers=headers, params=params)
    data = response.json()

    flights = {}
    for quote in data['Quotes']:
        price = quote['MinPrice']
        origin_airport_name = next((d['Name'] for d in data['Places'] if d['PlaceId'] == quote['OutboundLeg']['OriginId']), None)
        destination_airport_name = next((d['Name'] for d in data['Places'] if d['PlaceId'] == quote['OutboundLeg']['DestinationId']), None)
        company_name = next((d['Name'] for d in data['Carriers'] if d['CarrierId'] == quote['OutboundLeg']['CarrierIds'][0]), None)
        flight = {
            'price': price,
            'origin_airport_name': origin_airport_name,
            'destination_airport_name': destination_airport_name,
            'company_name': company_name,
        }
        if price in flights:
            flights[price].append(flight)
        else:
            flights[price] = [flight]

    return flights
```

```python
api_key = 'prtl6749387986743898559646983194'
start_date = '2023-04-01'
end_date = '2023-04-07'
origin_airport = 'SFO-sky'
destination_airport = 'LHR-sky'
flights = get_flight_prices(start_date, end_date, origin_airport, destination_airport, api_key)
```

Example of the API of league of legends

```python
# Functtion to get the item and cost and make 
def filter_items_by_cost(item_data, cost):
    # Create a dictionary of items, where the keys are the item IDs
    items = {}
    for item_id, item_info in item_data["data"].items():
        item_cost = item_info["gold"]["total"]
        if item_cost > cost:
            items[item_id] = {
                "name": item_info["name"],
                "description": item_info["description"],
                "cost": item_cost,
                "stats": item_info.get("stats", {}),
            }

    # Convert the items dictionary to a dataframe
    df = pd.DataFrame.from_dict(items, orient="index")

    return df
```

```python
# Define the API URL and retrieve the item data
version = "11.6.1"
language = "en_US"
url = f"https://ddragon.leagueoflegends.com/cdn/{version}/data/{language}/item.json"
response = requests.get(url)
item_data = response.json()
item_data
```

```python

# Filter the items by cost and print the resulting dataframe
df = filter_items_by_cost(item_data, 2300)
df.head()
```
